```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
#filtering out the dataset having only good or premium diamonds and diamonds of only D,F or J colors
#and taking only 100 observations

diamonds=diamonds[(diamonds.cut.isin(['Premium','Good']) & diamonds.color.isin(['D','F','J']))].sample(n=100,random_state=22)
diamonds.head()
```

```python
sns.scatterplot(x='carat',y='price',data=diamonds)
```

```python
#Alternatively
sns.scatterplot(x='x',y='y',data=diamonds)
```

```python
#To change the size of the dots:
sns.scatterplot(x='carat',y='price',s=100,data=diamonds)
```

```python
#To change the type of symbol being used:
sns.scatterplot(x='carat',y='price',marker='*',s=100,data=diamonds)
```

```python
#we can also outline the markers if we want:
sns.scatterplot(x='carat',y='price',edgecolor='black',data=diamonds)
```

```python
#also, the transparency of the dots can be varied to increase the visibility of overlapping dots
sns.scatterplot(x='carat',y='price',size='cut',alpha=0.5,data=diamonds)
```

Categorising data to see another variable symbolised onn the plot via different
coloured,shaped(styled) or sized dots

```python
#HUE property
sns.scatterplot(x='carat',y='price',hue='cut',data=diamonds)
```

```python
#to use some other colours
sns.scatterplot(x='carat',y='price',hue='cut',data=diamonds,palette=['r','g','b','k','y'])
```

```python
#Instead of using categorical values for hue, we can also plot using variables having
#numerical values,for eg: the depth column in diamonds
sns.scatterplot(x='carat',y='price',hue='depth',data=diamonds)
```

```python
#SHAPE/STYLE Property
sns.scatterplot(x='carat',y='price',style='cut',data=diamonds)
```

```python
#SIZE Property
sns.scatterplot(x='carat',y='price',size='cut',data=diamonds)
plt.show()
#we can also customise the sizes to be used:
sns.scatterplot(x='carat',y='price',size='cut',sizes=[20,150,150,20,20],data=diamonds)
plt.show()
```

```python
#similar to hue, we can also use the size to plot variables with numeric values
sns.scatterplot(x='carat',y='price',size='depth',data=diamonds)
```

```python
#Also, we can use all 3 properties simultaneously to represent the same variable, making the
#data even more readable
sns.scatterplot(x='carat',y='price',size='cut',hue='cut',style='cut',data=diamonds)
```

```python
#Alternatively, we could also use different properties to represent different variables, hence displaying more information
sns.scatterplot(x='carat',y='price',hue='cut',style='color',data=diamonds)
```
