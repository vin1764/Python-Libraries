```python
#DISTRIBUTION PLOT
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
cars=sns.load_dataset('mpg')
cars.dropna()
```

```python
#a default distplot consists of a combination of a kde plot and a histogram
sns.distplot(cars.weight)
```

```python
#To also see the rugplot in the distplot
sns.distplot(cars.weight,rug=True)
```

```python
#To change the number of bins in the histogram of the distplot
sns.distplot(cars.weight,bins=20)
plt.show()
#Alternatively,we can also see customised bins by entering a list of bins
sns.distplot(cars.weight,bins=[1500,2000,2500,3500,5200])
plt.show()
```

```python
#To change the transparency of the histogram in the distplot
sns.distplot(cars.weight,kde=False,hist_kws={'alpha':1}) #kde=Faalse hides the kdeplot and show only the histogram
```

```python
#Types of histograms
sns.distplot(cars.weight,kde=False,hist_kws={'histtype':'step'})
plt.show()
sns.distplot(cars.weight,kde=False,hist_kws={'histtype':'stepfilled'})
plt.show()
```
