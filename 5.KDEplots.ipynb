```python
#KERNEL DENSITY ESTIMATION PLOTS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars=sns.load_dataset('mpg')
cars.dropna()
```

```python
#To make a kde plot of horsepower
sns.kdeplot(cars.horsepower)
```

```python
#To shade the area under the kde plot
sns.kdeplot(cars.horsepower,fill=True)
```

```python
#To change the bandwidth of the kde plot
sns.kdeplot(cars.horsepower,bw_adjust=1) #increasing the bandwidth smoothens the curve,while decreasing it increases the number of peaks
plt.show()                               #default bw=0.5
import ipywidgets as widgets
def bandwidth_widget(bw=0.5):
    sns.kdeplot(cars.horsepower,lw=3,fill=True,bw_adjust=bw)
    plt.xlim=(0,250)
    plt.ylim=(0,0.016)
    
widgets.interact(bandwidth_widget,bw=(0.1,3)) #here 0.1 is the minimum value of the slider, and 3 is the maximum value that the slider can take.
                                              #Also, since this is a float slider, the default step is 0.1, but that can be changed
```

```python
#BIVARIATE KDE PLOTS
#To plot a kde plot between 2 series of a data

sns.kdeplot(cars.horsepower,cars.mpg)
```

```python
#To see more rings of the plot and to see a shaded plot
sns.kdeplot(cars.horsepower,cars.mpg,n_levels=20,fill=True)
```

```python
#Also, to see the densities represented by each color
sns.kdeplot(cars.horsepower,cars.mpg,n_levels=20,
            fill=True,
            cbar=True) #cbar stands for color bar
```
