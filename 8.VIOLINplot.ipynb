```python
#Violin plots are basically a combination of kde and box plots
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
cars=sns.load_dataset('mpg')
cars.dropna()
cars=cars[cars.cylinders.isin([4,6,8])]
cars.head()
```

```python
sns.violinplot(cars.displacement)
```

```python
#categorical data on violin plot
sns.violinplot(x=cars.cylinders,y=cars.displacement,hue=cars.origin)
```

```python
#Comparing two datas on the same violin plot
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True)
```

```python
#We can also change how the inner part of the split violins look so that each side represents the 25, 50 and 75% datas
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,inner='quartiles')
```

```python
#Normally in a violin split plot, the area of the 2 datasets is the same, which makes it confusing to determine which value
#has a higher count. We can scale the chart to represent the count of the 2 datas
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,scale='count')
```

```python
#Here, seaborn first split our data according to 4&6 cylinders and then according to the origin. To see the actual count comparison
#according to the number of cylinders as well as the origin:
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,scale='count',scale_hue=False)
#Here we can notice the number of 4 cylinder cars is way more than the number of 6 cylinder cars
```

All the styling options in boxplot also apply to violin plots
