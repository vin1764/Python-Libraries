Cost function in logistic regression is given as:

Cost=(1/m)*summation(L(f_wb,y))

where L=Loss i.e the cost of one training example


L(f_wb_i,y_i)  

=   -log(f_wb_i)        when y_i=1

=   -log(1-f_wb_i)      when y_i=0



=>L(f_wb_i,y_i) = [-y_i*log(f_wb_i)]-[(1-y_i)*log(1-f_wb_i)]
               

```python
import numpy as np
import matplotlib.pyplot as plt 
from plt_logistic_loss import  plt_logistic_cost, plt_two_logistic_loss_curves, plt_simple_example
from lab_utils_common import  plot_data, sigmoid, dlc
```

```python
#Cost Function

def compute_logistic_cost(x,y,w,b):
    m=len(x)
    cost=0
    
    for i in range(m):
        z_i=np.dot(w,x[i])+b
        f_wb_i=sigmoid(z_i)
        cost=cost+(-y[i]*np.log(f_wb_i)-(1-y[i])*(np.log(1-f_wb_i)))
        
    cost=cost/m
    
    return cost
```

Lets plot w0=1,w1=1,b=-3 and w0=1,w1=1,b=-4 on a graph and check which fits better, and then compare their costs



```python
x_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1]) 

x0=np.arange(0,6)
x1=3-x0
x1_other=4-x0

fig,ax=plt.subplots(1,1,figsize=(4,4))
#Plotting the decision boundaries
ax.plot(x0,x1, c=dlc["dlblue"], label="b=-3")
ax.plot(x0,x1_other, c=dlc["dlmagenta"], label="b=-4")
ax.axis([0, 4, 0, 4])

plot_data(x_train,y_train,ax)
ax.set_xlabel('x0')
ax.set_ylabel('x1')
plt.legend()
plt.show()
```

We can see that b=-4 is a bad model. Now checking this with the cost function

```python
w=np.array([1,1])
print(f'Cost at b=-3: {compute_logistic_cost(x_train,y_train,w,b=-3)}')
print(f'Cost at b=-4: {compute_logistic_cost(x_train,y_train,w,b=-4)}')
```

We can see that the cost for b=-4 is more implying that it is a worse model
