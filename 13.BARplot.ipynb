```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.groupby('species').body_mass_g.mean()
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins)
```

```python
#to make a horizontal barplot
sns.barplot(x='body_mass_g',y='species',data=penguins)
```

```python
#we can control the length of the confidence intervals shown in the graph.By default, ci=95
sns.barplot(x='species',y='body_mass_g',data=penguins,ci=50)
```

```python
#instead of having the mean, we can also have other numeric properties plotted on the bar plot.eg:median,standard deviation etc
sns.barplot(x='species',y='body_mass_g',data=penguins,estimator=np.std)
```

CATEGORICAL VARIABLES

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,hue='sex')
```

```python
#we can also change the order of the categories and customise it ourselves
sns.barplot(x='species',y='body_mass_g',data=penguins,hue='sex',order=['Adelie','Gentoo','Chinstrap'],hue_order=['Female','Male'])
```

```python
#To change the colors of bars
sns.barplot(x='species',y='body_mass_g',data=penguins,color='cyan',errcolor='red',errwidth=8)
```

```python
#instead of giving the same colour to all bars, we can instead pass in a palette of colors
sns.barplot(x='species',y='body_mass_g',data=penguins,palette='cool',edgecolor='black',lw=10)
```
