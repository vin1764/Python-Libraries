```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

tips=sns.load_dataset('tips')
print(tips.head())
```

```python
sns.pairplot(tips)
```

```python
#as we can see, the default pairplot will have a histogram on the diagonal plots
#and a scatterplot on the other plots. We can change this:
sns.pairplot(tips,diag_kind='kde',kind='reg')
```

```python
#TYPE OF PAIRPLOT
type(sns.pairplot(tips))#we can see that the type of this pairplot is a pairgrid
```

```python
#we can actually save a PAIRPLOT and then make changes to it later
#let say we want to add a kdeplot in all of the upper triangle plots of a pairplot g
g=sns.pairplot(tips)
g.map_upper(sns.kdeplot,n_levels=6,color='xkcd:wine red')
plt.show()
```

```python
#lets create a new column which has a yes for the weekends and no for the weekdays
tips['weekend']=tips.day.isin(['Sat','Sun'])
tips.head()
```

```python
#now,we can add categories in the form of hues in the pairplot
sns.pairplot(tips,hue='weekend')
```

```python
#let say we wanna see any 1 or 2 specific columns
sns.pairplot(tips,hue='weekend',vars=['total_bill','weekend'])
```

```python
#if we only want specific columns along the x and y axes
sns.pairplot(tips,hue='weekend',x_vars=['total_bill','size'],y_vars=['tip'])
```

```python
#we can change the size of the pairplots interactively (interactive widget will run in notebook)
from ipywidgets import interactive
def size_widget(height=2.5,aspect=1):
    sns.pairplot(tips,height=height,aspect=aspect)
    
interactive(size_widget,height=(1,3.5,0.5),aspect=(0.5,2,0.25))
```

```python
#To change the color palette
sns.pairplot(tips,hue='day',palette='plasma')
```

```python
#to change the properties of only specific plots of the pairplots:
sns.pairplot(tips,diag_kws={'color':'gray'})
plt.show()
sns.pairplot(tips,kind='reg',plot_kws={'color':'xkcd:salmon','ci':None})
plt.show()
#here we can also change the color of only the scatterpoints iw=f we want
sns.pairplot(tips,kind='reg',plot_kws={'color':'xkcd:salmon','ci':None,'scatter_kws':{'color':'blue'}})
plt.show()
```
