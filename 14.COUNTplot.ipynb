```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
#narrowing down the data
diamonds=diamonds[diamonds.clarity.isin(['SI1','VS2'])]
diamonds.shape
```

```python
sns.countplot(x='color',data=diamonds)
```

```python
diamonds.color.value_counts(sort=False) #this is the data plotted by the above countplot
```

```python
#we can also plot a horizontal countplot
sns.countplot(y='color',data=diamonds)
```

```python
#if we want to arrange the bars in ascending or descending order
sns.countplot(y='color',data=diamonds,order=diamonds.color.value_counts().index) #descending
plt.show()
sns.countplot(y='color',data=diamonds,order=diamonds.color.value_counts().index[::-1]) #ascending
plt.show()
```

```python
#Categorical variables

sns.countplot(x='color',data=diamonds,hue='clarity')
plt.show()
#note that here we can see all the clarity categories in the legend. If we want to see only specific categories
sns.countplot(x='color',data=diamonds,hue='clarity',hue_order=['SI1','VS2'])
plt.show()
```
