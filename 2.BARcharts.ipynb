```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.style  as style
style.use('ggplot')
```

```python
plt.clf()  #clf=clear figure, used to clear the plot and make it blank
```

```python
x=np.arange(4)   # x=[0,1,2,3]
y=[8.8,5.2,3.6,5.9]
plt.xticks(x,['Ankit','Hans','Joe','Flaco'])  #assigning names to the 0,1,2,3 values of the array x
plt.bar(x,y)
```

```python
#to change the colour of the bars:
x=np.arange(4)   # x=[0,1,2,3]
y=[8.8,5.2,3.6,5.9]
plt.xticks(x,['Ankit','Hans','Joe','Flaco'])  #assigning names to the 0,1,2,3 values of the array x
plt.bar(x,y,color='y')
```

```python
#To get all different coloured bars:
x=np.arange(4)   # x=[0,1,2,3]
y=[8.8,5.2,3.6,5.9]
plt.xticks(x,['Ankit','Hans','Joe','Flaco'])  #assigning names to the 0,1,2,3 values of the array x
plt.bar(x,y,color=['lime','r','k','tan'])
```
