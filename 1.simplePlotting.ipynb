```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.style  as style
style.use('ggplot')
```

```python
#simple plot with 4 numbers

plt.plot([1,2,3,4])
plt.show()

plt.plot([1,3,2,4])
plt.show()
```

```python
#Plotting x and y values and naming the graph and the axes
plt.plot([1,2,3,4],[1,4,9,16])
plt.title('Test Plot',fontsize=8,color='g')
plt.xlabel('Number n')
plt.ylabel('n^2')
```

```python
#To change the figure size of the plot and print dots instead of lines
plt.figure(figsize=(2,5)) #2 inches wide, 5 inches tall
plt.plot([1,2,3,4],[1,4,9,16],'ro') #ro means to print red coloured(r) dots(o) instead of joining by lines
```

```python
#To set axis scales:: x:0-6 and y:0-20
plt.axis([0,6,0,20]) #[xmin,xmax,ymin,ymax]
```

```python
#To write text somewhere on the plot:
plt.plot([1,2,3,4],[1,4,9,16])
plt.annotate('(1,1)',(1,1))
plt.annotate('(2,4)',(2,4))
plt.annotate('(3,9)',(3,9))
plt.annotate('(4,16)',(4,16))
```

```python
#Plotting sets of 10 random points
d={'Red O': np.random.rand(10),
   'Grn X': np.random.rand(10)}

df=pd.DataFrame(d)
df.plot(style=['ro','gx'])
```

```python
#plotting random dots in a scatter diagram
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
sizes=(30*np.random.rand(N))**2  #0 to 15 point radii
plt.scatter(x,y,s=sizes,c=colors,alpha=0.5)  #alpha= transparenecy of dots
```
