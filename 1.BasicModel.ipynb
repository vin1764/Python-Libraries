For Logistic Regression, we need to classify datapoints. Let say we have to classify based on various features whether an animal is a cat or not. So, for this we use a sigmoid function in our linear regression model since the result of a logistic regression can only be 0(false) or 1(true)

```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_common import draw_vthresh,plot_data
```

```python
#Sigmoid Function, g(z)=1/(1+e^(-z))

def sigmoid(z):
    g=1/(1+np.exp(-z))
    
    return g
```

```python
print(f'sigmoid(0)={sigmoid(0)}')
print(f'sigmoid(1)={sigmoid(1)}')

print(f'sigmoid of a very large negative number={sigmoid(-9999)}')
print(f'sigmoid of a very large positive number={sigmoid(9999)}')
```

```python
z=np.arange(-10,11)
fig,ax=plt.subplots(1,1,figsize=(5,3))
ax.plot(z,sigmoid(z))
ax.set_title('Sigmoid Plot')
ax.set_xlabel('z')
ax.set_ylabel('Sigmoid(z)')
draw_vthresh(ax,0)
plt.show()
```

The logistic regression model has the form
f_wb=sigmoid(w.x+b)=1/(1+e^-(w.x+b))



```python
x=np.array([[0.5,1.5],[1,1],[1.5,0.5],[3,0.5],[2,2],[1,2.5]])
y=np.array([0,0,0,1,1,1]).reshape(-1,1)

fig,ax=plt.subplots(1,1,figsize=(4,4))
ax.axis([0,4,0,3.5])
plot_data(x,y,ax)
ax.set_xlabel('$x_0$')
ax.set_ylabel('$x_1$')
plt.show()
```

Now the training model is:

f_w0,w1,b=sigmoid(w0x0+w1x1+b)



From sigmoid plot, sigmoid(z)>0.5 for z>0


  if $\mathbf{w} \cdot \mathbf{x} + b >= 0$, the model predicts $y=1$
  
  if $\mathbf{w} \cdot \mathbf{x} + b < 0$, the model predicts $y=0$




So, for (f_w0,w1,b)=1, z=(w0x0+w1x1+b)>0


let say w0=w1=1, b=-3

=>     x0+x1>3 ---------------->DECISION BOUNDARY

```python
x0 = np.arange(0,6)
x1 = 3 - x0

# Plot the decision boundary

fig,ax = plt.subplots(1,1,figsize=(5,4))
ax.plot(x0,x1,c="b")
ax.axis([0, 4, 0, 3.5])

ax.fill_between(x0,x1, alpha=0.8)

# Plot the original data
plot_data(x,y,ax)
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
plt.show()
```

* Any point in the shaded region (under the line) is classified as $y=0$.  Any point on or above the line is classified as $y=1$. This line is known as the "decision boundary".
