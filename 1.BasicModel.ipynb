```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating 2 training data points
x_train=np.array([1,2])
y_train=np.array([300,500])
print(f'x_train={x_train}')
print(f'y_train={y_train}')

m=len(x_train)
print(f'Number of training examples = {m}')
```

```python
i=0
x_i=x_train[i]
y_i=y_train[i]
print(f'x_{i}={x_i}\ny_{i}={y_i}')
```

```python
plt.scatter(x_train,y_train)
plt.title('House Prices')
plt.xlabel('Area of House(in 1000sqft)')
plt.ylabel('Price of House(in 1000$)')
plt.show()
```

```python
#defining parameters
w=100
b=100
print(f'w={w}\nb={b}')
```

```python
#defining the function to create the model
def compute_model_output(x,w,b):
    m=len(x)
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
        
    return f_wb
```

```python
#plotting a graph of x against f_wb, i.e. the house area against the predicted cost assuming w and b to be 100

plt.plot(x_train,compute_model_output(x_train,w,b),label='Predicted Model')
plt.scatter(x_train,y_train,color='Red',label='Actual Data')
plt.title('Housing Prices')
plt.xlabel('Area of House(in 1000sqft)')
plt.ylabel('Price of House(in 1000$)')
plt.legend()
plt.show()
```

```python
#We can see that this line is poorly fitted with the data because of the w and b choices
#Mathematically, the best choices for w and b will be w=200,b=100. Using these values
plt.plot(x_train,compute_model_output(x_train,200,100),label='Predicted Model')
plt.scatter(x_train,y_train,color='Red',label='Actual Data')
plt.title('Housing Prices')
plt.xlabel('Area of House(in 1000sqft)')
plt.ylabel('Price of House(in 1000$)')
plt.legend()
plt.show()
```
