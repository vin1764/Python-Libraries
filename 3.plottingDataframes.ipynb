```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style  as style
style.use('ggplot')
```

```python
df=pd.read_csv('Fremont_weather.txt')
df.head()
```

```python
plt.bar(df.month,df.record_high,color='r') #To plot a bar chart with x axis=months and y axis= record_high,bar color=red
plt.bar(df.month,df.record_low, color='c') #To plot a bar chart on the same graph with x=month, y=record_low,bar color= cyan
plt.plot(df.month,df.avg_high, color='k') #To plot a line chart on the same graph
plt.plot(df.month,df.avg_low, color='b') #To plot another line chart on the graph
plt.legend() # To show the scale
```

```python
#SUBPLOTS

fig=plt.figure()
plt.suptitle('My Subplots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222,facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224)
```

```python
fig=plt.figure()
fig,plots=fig.subplots(2,sharex=True) #sharex=true means they will share the x axis
fig.suptitle('Sharing X axis')
x=range(0,200,5)
y=[n**0.8 for n in x]
plots[0].plot(x,y,color='r')
plots[1].scatter(x,y)
```

```python
#To save the figure as a image file
plt.figure(figsize=(4,3),dpi=100) #dpi is the resolution setting
plt.plot([245,170,148,239,161,196,112,258])
plt.axis([0,7,0,300])
plt.title('Flight Data')
plt.xlabel('Speed')
plt.ylabel('No. of Flights')
plt.savefig('Flights.png')
```
